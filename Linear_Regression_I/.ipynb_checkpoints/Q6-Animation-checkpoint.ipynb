```python
import autograd.numpy as np
import matplotlib.pyplot as plt
from autograd import grad
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
import cv2
```

```python
def gradientDescentRegression(X,y,alpha = 0.1):
    
    size_x,size_y = np.shape(X)
    X_ = np.ones([size_x,size_y+1])
    X_[:,1:] = X
    theta = np.zeros(len(X_[0]))
    for i in range(100): 
        gr = gradient(X_,y,theta)
        theta = theta-gr*alpha

    
    return theta
```

```python
def error(x,y,theta):
    
    err = y - sum(x*theta)
    return err
```

```python
def f(th0,th1):

    residue = (6-th0-th1)**2+(10-th0-3*th1)**2 + (16-th0-6*th1)**2
    return residue
```

```python
#Q6(a) Contour Plot of Residual sum of squares. 
th0 = np.arange(-20.0, 20.0, 1)
th1 = np.arange(-30.0, 30.0, 1)

X, Y = np.meshgrid(th0, th1)

Z = (6-X-Y)**2+(10-X-3*Y)**2 + (16-X-6*Y)**2

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)


ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Countour plot of residual sum of squares')
ax.set_xlabel("theta0")
ax.set_ylabel("theta1")




```

```python
def auto_cost(theta):
    pred = np.dot(X_auto, theta)
    return np.sqrt(((pred - Y_auto) ** 2).mean(axis=None))
```

```python
def gradient(X,y,theta): 
    rows = len(X)

    grad = np.zeros(len(theta))
    for j in range(len(theta)): 
        for i in range(len(X)):

            grad[j]+=-2*X[i,j]*error(X[i],y[i],theta)
    return 1.0*grad/rows
            
```
