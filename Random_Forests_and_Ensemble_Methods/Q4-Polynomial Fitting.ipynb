```python
from numpy import *
import random as rnd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating random data points. 
def rand(number,m,c):
    x = linspace(0,1,number)
    y = m*x+c+0.8*random.rand(number)
    return x,y
```

```python
def bagging(x,y):
    left = int(5*len(x)/10)
    rand_list = []
    for item in range(left): 
        index = rnd.randrange(len(x))
        rand_list.append(index)
    x_list = [x[i] for i in rand_list]
    y_list = [y[i] for i in rand_list]
    return x_list,y_list
    
```

```python
m  = 2
c = 5
no = 50
```

```python
#Q4(A) Creating Random Data Points. 
x,y = rand(no,m,c)
plt.plot(x,y)
```

```python
#Q4(B) Applying 5 degree polynomial fit on data
p = poly1d(polyfit(x, y, 5))
t = linspace(0,1,50)
plt.plot(x, y, 'o', t, p(t), '-')
plt.show()
```

```python
#Q4(C) Averaging multiple curves.

p = zeros(6)
for i in range(1000):
    x_up, y_up  = bagging(x,y)
    p1 = poly1d(polyfit(x_up,y_up,5))
    parray = asarray(p1)
    p+=parray

p = np.poly1d(p/1000)

t = linspace(0.01,0.99,50)
plt.plot(x, y, 'o', t, p(t), '-')
plt.show()
```
